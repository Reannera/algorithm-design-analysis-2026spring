idea:

````python
# add your code here
"""
算法思想：
题目给出的伪代码对应 Stooge Sort。其基本做法不是把区间一分为二，而是反复处理两个互相重叠的 2/3 区间。
对于当前区间 arr[left...right]，先递归排序前 2/3，再递归排序后 2/3，最后再次递归排序前 2/3。
由于两个递归区间存在重叠，较大的元素会被逐步移动到右端，较小的元素会被逐步移动到左端，最终整个区间有序。
该方法能够完成排序，但递归调用次数较多，因此效率较低。

时间复杂度：
递推式为 T(n) = 3T(2n/3) + O(1)，因此时间复杂度约为 O(n^2.71)。
"""
import math


def do_something(arr, left, right):
    n = right - left + 1

    if n <= 1:
        return

    if n == 2:
        if arr[left] > arr[right]:
            arr[left], arr[right] = arr[right], arr[left]
        return

    part = math.ceil(2 * n / 3)

    # 对前 2/3 区间排序
    do_something(arr, left, left + part - 1)
    # 对后 2/3 区间排序
    do_something(arr, right - part + 1, right)
    # 再次对前 2/3 区间排序
    do_something(arr, left, left + part - 1)


nums = [5, 3, 1, 4, 2]
do_something(nums, 0, len(nums) - 1)
print(nums)
# your algorithm time complexity is: O(n^2.71)
````

## 问题 3

给定一个包含n个元素的数组 `profits` , 它的第 `i` 个元素 `profits[i]` 表示一支股票第 `i` 天的**收益**（正数表示涨, 负数表示跌）。你只能选择 **某一天** 买入这只股票, 并选择在 **未来的某一个不同的日子** 卖出该股票。

1. 设计一个算法来计算你所能获取的最大利润和对应买入和卖出的日期。请分析算法方案, 计算其时间复杂度, 并且使用python编程实现该算法。

2. \* 设计一个时间复杂度为 $O(n)$的算法实现该算法

e.g. :
---
profits=[3,2,1,-7,5,2,-1,3,-1], 第5天买入, 第8天卖出, 收益最大：9

idea:

````python
# add your code here
"""
算法思想：
该问题可以转化为最大连续子数组和问题。
设 cur_sum 表示“以第 i 天作为卖出日时，能够得到的最大累计收益”，其含义是当前连续区间对答案的贡献。
如果 cur_sum 小于 0，则说明此前这段区间只会拖累后续结果，因此应舍弃此前缀，从当前位置重新开始计算。
如果 cur_sum 大于等于 0，则说明此前区间仍然有贡献，可以继续向后扩展。
在扫描过程中同步记录当前区间起点以及历史最优区间的起止位置，即可得到最大收益及对应日期。

时间复杂度：
只需一次线性扫描，时间复杂度为 O(n)。
"""
def max_profit(profits):
    n = len(profits)
    if n == 0:
        return 0, -1, -1

    cur_sum = profits[0]
    best_sum = profits[0]
    start = 0
    best_start = 0
    best_end = 0

    for i in range(1, n):
        # 当前累计收益为负时，从第 i 天重新开始
        if cur_sum < 0:
            cur_sum = profits[i]
            start = i
        else:
            cur_sum += profits[i]

        if cur_sum > best_sum:
            best_sum = cur_sum
            best_start = start
            best_end = i

    return best_sum, best_start + 1, best_end + 1


profits = [3, 2, 1, -7, 5, 2, -1, 3, -1]
profit, buy_day, sell_day = max_profit(profits)

print("最大收益:", profit)
print("买入日期:", buy_day)
print("卖出日期:", sell_day)
# your algorithm time complexity is: O(n)
````

## 问题 4

给定一个包含不同整数元素的数组 $ A[1..n]$ ,并且满足条件：$A[1]>A[2]$ 并且 $A[n-1]<A[n]$; 规定：如果一个元素比它两边的邻居元素都小, 即：$A[x]<A[x-1], A[x]<A[x+1]$ , 称这个元素A[x]为“局部最小”。通过遍历一次数组, 我们可以很容易在 $O(n)$的时间复杂度下找到一个局部最小值, 


1. 分析该问题, 设计一个算法在$O(logn)$的时间复杂度下找到一个局部最小(返回数值), 要求：分析算法设计思路, 并且使用python编程实现
2. \* 设计算法找出所有局部最小值, 分析算法设计思路, 并使用python编程实现

e.g.:
---
A=[9, 3, 7, 2, 1, 4, 5 ] 时,  局部最小元素为 3, 1

idea:

````python
# add your code here
"""
1. 求一个局部最小值
算法思想：
由题设 arr[0] > arr[1] 且 arr[n - 2] < arr[n - 1] 可知，数组左端呈下降趋势，右端呈上升趋势，
因而中间至少存在一个局部最小值。
二分时比较 arr[mid] 与 arr[mid + 1]：
若 arr[mid] > arr[mid + 1]，说明 mid 右侧仍在下降，局部最小值必定位于右半部分；
否则说明当前位置之后开始上升，局部最小值位于左半部分或 mid 本身。
因此可以像二分查找一样不断缩小搜索区间，最终定位到一个局部最小值。

时间复杂度：
每次将搜索区间缩小一半，因此时间复杂度为 O(log n)。
"""
def find_one_local_min(arr):
    n = len(arr)

    if n < 3:
        raise ValueError("数组长度至少要有 3 个数")

    if not (arr[0] > arr[1] and arr[n - 2] < arr[n - 1]):
        raise ValueError("数组不满足题目给定的条件")

    left = 1
    right = n - 2

    while left < right:
        mid = (left + right) // 2

        if arr[mid] > arr[mid + 1]:
            left = mid + 1
        else:
            right = mid

    return arr[left]


nums = [9, 3, 7, 2, 1, 4, 5]
print(find_one_local_min(nums))

"""
2. 求所有局部最小值
算法思想：
当要求输出全部局部最小值时，答案可能不止一个，此时无法继续使用二分策略。
只需顺序扫描数组的中间位置，并检查每个元素是否同时小于左右相邻元素。
满足 arr[i] < arr[i - 1] 且 arr[i] < arr[i + 1] 的元素即为局部最小值，将其加入结果即可。

时间复杂度：
需要扫描整个数组一次，因此时间复杂度为 O(n)。
"""
def find_all_local_min(arr):
    ans = []

    for i in range(1, len(arr) - 1):
        if arr[i] < arr[i - 1] and arr[i] < arr[i + 1]:
            ans.append(arr[i])

    return ans


nums = [9, 3, 7, 2, 1, 4, 5]
print(find_all_local_min(nums))
# your algorithm time complexity is: O(log n), O(n)
````

## 问题 5

给定包含n个不同数字的一组数, 寻找一种基于比较的算法在这组数中找到k个最小的数字, 并按顺序输出它们。

1. 将n个数先进行排序, 然后按顺序输出最小的k个数。要求：选择合适的排序算法实现上述操作, 计算算法时间复杂度, 并使用python编程实现。
2. 建立一个包含这n个数的堆（heap）, 并且调用 k 次Extract-min 按顺序输出最小的k个数。使用往空堆中不断插入元素的方法建立堆, 分析这种方法建堆的时间复杂度, 并使用python编程实现
3. \* 假设数组中包含的数据总数目超过了计算机的存储能力, 请设计一个算法, 找到这堆数据的前k小的数值, 计算时间复杂度, 并使用python实现该算法, 假设计算机一定能存储k个数据。

e.g.：
---
数组arr=[5,4,3,2,6,1,88,33,22,107] 的前3个最小数据为：1, 2, 3

idea:

````python
# add your code here
"""
1. 先排序，再输出前 k 个最小值
算法思想：
该方法首先将全部元素排序，再从有序序列中取出前 k 个元素。
排序完成后，数组中的相对大小关系已经完全确定，因此前 k 个位置上的元素必然是全局最小的 k 个元素。
此处采用归并排序实现，原因在于归并排序属于基于比较的稳定排序算法，时间复杂度较为明确。

时间复杂度：
排序部分为 O(n log n)，取前 k 个元素为 O(k)，总时间复杂度为 O(n log n)。
"""
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    return merge(left, right)


def merge(left, right):
    result = []
    i = 0
    j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    while i < len(left):
        result.append(left[i])
        i += 1

    while j < len(right):
        result.append(right[j])
        j += 1

    return result


def find_k_smallest_by_sort(arr, k):
    sorted_arr = merge_sort(arr)
    return sorted_arr[:k]


arr = [5, 4, 3, 2, 6, 1, 88, 33, 22, 107]
k = 3
print(find_k_smallest_by_sort(arr, k))

"""
2. 建立最小堆并执行 k 次 extract-min
算法思想：
最小堆具有“堆顶元素始终最小”的性质，因此可以利用该性质逐次取出最小值。
先将全部元素逐个插入最小堆，建堆完成后，连续执行 k 次 extract-min，依次得到当前最小元素。
由于每次删除堆顶后堆结构仍会被调整为最小堆，因此取出的前 k 个元素天然按从小到大的顺序给出。

时间复杂度：
建堆过程为 O(n log n)，执行 k 次 extract-min 为 O(k log n)，总时间复杂度为 O(n log n + k log n)。
"""
class MinHeap:
    def __init__(self):
        self.heap = []

    def insert(self, x):
        self.heap.append(x)
        self._sift_up(len(self.heap) - 1)

    def extract_min(self):
        if not self.heap:
            return None

        min_value = self.heap[0]
        last_value = self.heap.pop()

        if self.heap:
            self.heap[0] = last_value
            self._sift_down(0)

        return min_value

    def _sift_up(self, index):
        while index > 0:
            parent = (index - 1) // 2
            if self.heap[parent] <= self.heap[index]:
                break
            self.heap[parent], self.heap[index] = self.heap[index], self.heap[parent]
            index = parent

    def _sift_down(self, index):
        n = len(self.heap)

        while True:
            left = 2 * index + 1
            right = 2 * index + 2
            smallest = index

            if left < n and self.heap[left] < self.heap[smallest]:
                smallest = left
            if right < n and self.heap[right] < self.heap[smallest]:
                smallest = right
            if smallest == index:
                break

            self.heap[index], self.heap[smallest] = self.heap[smallest], self.heap[index]
            index = smallest


def find_k_smallest_by_min_heap(arr, k):
    heap = MinHeap()

    for x in arr:
        heap.insert(x)

    result = []
    for _ in range(k):
        result.append(heap.extract_min())

    return result


print(find_k_smallest_by_min_heap(arr, k))

"""
3. 数据规模超过内存容量，仅能存储 k 个元素
算法思想：
当全部数据无法一次性存入内存时，需要在扫描数据流的过程中实时维护答案。
为此可以构造一个大小固定为 k 的最大堆，使堆中始终保存“当前见过的最小 k 个元素”。
此时堆顶元素是这 k 个元素中的最大值，因此新读入元素只有在小于堆顶时才有资格进入结果集合。
每次更新后，最大堆仍然保持了“堆中元素即为当前最小 k 个元素”的不变性。

时间复杂度：
堆的规模始终为 k，因此总时间复杂度为 O(n log k)。
"""
import heapq


def find_k_smallest_large_data(data_stream, k):
    max_heap = []

    for x in data_stream:
        if len(max_heap) < k:
            heapq.heappush(max_heap, -x)
        else:
            current_max = -max_heap[0]
            if x < current_max:
                heapq.heapreplace(max_heap, -x)

    result = [-x for x in max_heap]
    result.sort()
    return result


print(find_k_smallest_large_data(arr, k))
# your algorithm time complexity is: O(n log n), O(n log n + k log n), O(n log k)
````

## 问题 6

> 星球碰撞问题：现有n个星球，在同一条直线上运行，如数组A所示，元素的绝对值表示星球的质量，负数表示星球自右向左运动，正数表示星球自左向右运动，当两个星球相撞的时候，质量小的会消失，大的保持不变，**质量相同的两个星球碰撞后自右向左运动的星球消失，自左向右的星球保持不变**，假设所有星球的速度大小相同。
>
> $ A=[23,-8, 9, -3, -7, 9, -23, 22] $

请设计一个算法模拟星球的运行情况，输出最终的星球存续情况（输出一个数组），分析算法设计思路，计算时间复杂度，并基于python编程实现。

e.g.
---
1.  输入： A=[-3,-6,2,8, 5,-8,9,-2,1]， 输出：[-3, -6, 2, 8, 9, 1]

2. 输入：A=[23,-8, 9, -3, -7, 9, -23, 22], 输出：[23, 22]

idea:


````python
# add your code here
"""
算法思想：
采用栈维护当前仍然存活的星球序列。
只有当栈顶星球向右运动而当前星球向左运动时，两者才可能迎面相撞，因此仅在这一情况下进行比较。
若当前星球质量更大，则栈顶星球被消去，当前星球继续与新的栈顶比较；
若当前星球质量更小，则当前星球被消去；若两者质量相同，则依据题意保留向右运动的星球。
通过这一过程，栈中始终保存当前已经处理部分的最终存活状态。

时间复杂度：
每个星球至多入栈和出栈各一次，因此时间复杂度为 O(n)。
"""
def asteroid_collision(planets):
    stack = []

    for planet in planets:
        alive = True

        while alive and stack and stack[-1] > 0 and planet < 0:
            if stack[-1] < -planet:
                stack.pop()
            elif stack[-1] == -planet:
                alive = False
            else:
                alive = False

        if alive:
            stack.append(planet)

    return stack


sample1 = [-3, -6, 2, 8, 5, -8, 9, -2, 1]
sample2 = [23, -8, 9, -3, -7, 9, -23, 22]

print(asteroid_collision(sample1))
print(asteroid_collision(sample2))
# your algorithm time complexity is: O(n)
````

## 问题 7

> 0-1背包问题：现在有4块大理石，每个大理石对应的重量和价值使用一个元组表示, 即（重量，价值），4块大理石的重量和价值为：[(5,10), (4,40), (6,30), (3, 50)], 假设你有一辆最大承重为9的小推车，请问使用这个小推车装大理石的最大价值为多少

请设计一个算法解决该问题，分析算法设计思路，计算时间复杂度，并基于python编程实现

idea:


````python
# add your code here
"""
算法思想：
该问题属于典型的 0-1 背包问题，其中每块大理石只有“选入”或“不选”两种决策。
定义 dp[i][c] 表示在只考虑前 i 块大理石且背包容量为 c 时能够取得的最大价值。
对于第 i 块大理石，若不选，则状态保持为 dp[i - 1][c]；若选入，则价值变为 dp[i - 1][c - weight] + value。
因此只需比较上述两种情况即可得到当前状态的最优值。
在全部状态计算完成后，再从后向前回溯 dp 表，即可恢复实际被选择的大理石集合。

时间复杂度：
共有 n 个物品、W 个容量状态，因此时间复杂度为 O(nW)。
"""
def knapsack_01(items, capacity):
    n = len(items)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        weight, value = items[i - 1]
        for c in range(capacity + 1):
            dp[i][c] = dp[i - 1][c]
            if c >= weight:
                dp[i][c] = max(dp[i][c], dp[i - 1][c - weight] + value)

    chosen = []
    c = capacity
    for i in range(n, 0, -1):
        if dp[i][c] != dp[i - 1][c]:
            chosen.append(items[i - 1])
            c -= items[i - 1][0]

    chosen.reverse()
    return dp[n][capacity], chosen


items = [(5, 10), (4, 40), (6, 30), (3, 50)]
capacity = 9

max_value, chosen_items = knapsack_01(items, capacity)
print(max_value)
print(chosen_items)
# your algorithm time complexity is: O(nW)
````

````python
# add your code here
"""
注：此题使用了大模型进行辅助
算法思想：
二叉搜索树的中序遍历序列应为严格递增序列，因此一旦有两个节点的值被交换，遍历过程中必然出现逆序现象。
若两个错误节点在中序序列中不相邻，则会出现两次逆序；若二者相邻，则只会出现一次逆序。
因此可以在中序遍历时维护前驱节点 prev：
第一次出现逆序时记录较大的前驱节点作为第一个错误节点，最后一次出现逆序时记录较小的当前节点作为第二个错误节点。
在遍历结束后交换这两个节点的值，即可恢复原二叉搜索树。

时间复杂度：
每个节点只遍历一次，因此时间复杂度为 O(n)，递归栈空间复杂度为 O(h)。
"""
from collections import deque


class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def build_tree(level_order):
    if not level_order:
        return None

    nodes = [None if value is None else TreeNode(value) for value in level_order]
    children = deque(nodes[1:])

    for node in nodes:
        if node is not None:
            if children:
                node.left = children.popleft()
            if children:
                node.right = children.popleft()

    return nodes[0]


def tree_to_list(root):
    if root is None:
        return []

    result = []
    queue = deque([root])

    while queue:
        node = queue.popleft()
        if node is None:
            result.append(None)
        else:
            result.append(node.val)
            queue.append(node.left)
            queue.append(node.right)

    while result and result[-1] is None:
        result.pop()

    return result


def recover_tree(root):
    first = None
    second = None
    prev = None

    def inorder(node):
        nonlocal first, second, prev
        if node is None:
            return

        inorder(node.left)

        if prev is not None and prev.val > node.val:
            if first is None:
                first = prev
            second = node

        prev = node
        inorder(node.right)

    inorder(root)

    if first is not None and second is not None:
        first.val, second.val = second.val, first.val

    return root


root1 = build_tree([1, 3, None, None, 2])
recover_tree(root1)
print(tree_to_list(root1))

root2 = build_tree([3, 1, 4, None, None, 2])
recover_tree(root2)
print(tree_to_list(root2))
# your algorithm time complexity is: O(n)
````

idea:


````python
# add your code here
"""
注：此题使用了大模型进行辅助
算法思想：
采用双指针方法从数组两端向中间扫描，并分别维护左右两侧已经出现过的最高柱子 left_max 和 right_max。
对于任意位置，其接水量取决于左右两侧较低的边界。
因此，当 height[left] <= height[right] 时，可以确定左侧当前位置的有效边界不超过右侧边界，
此时左侧位置的接水量仅由 left_max 决定；反之，右侧当前位置的接水量仅由 right_max 决定。
通过这种方式，每次都可以确定一侧当前位置的贡献，从而在线性时间内完成整体计算。

时间复杂度：
只需一次线性扫描，因此时间复杂度为 O(n)，额外空间复杂度为 O(1)。
"""
def trap(height):
    left = 0
    right = len(height) - 1
    left_max = 0
    right_max = 0
    water = 0

    while left < right:
        if height[left] <= height[right]:
            left_max = max(left_max, height[left])
            water += left_max - height[left]
            left += 1
        else:
            right_max = max(right_max, height[right])
            water += right_max - height[right]
            right -= 1

    return water


height = [0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1]
print(trap(height))
height = [4,2,0,3,2,5]
print(trap(height))
# your algorithm time complexity is: O(n)
````

设计一个算法计算从根节点到叶节点生成的所有数字之和 ，分析算法设计思路，计算时间复杂度，并基于python编程实现

叶节点:是指没有子节点的节点。

 输入：root = [1,2,3], 输出：25

idea:


````python
# add your code here
"""
算法思想：
采用深度优先搜索遍历每一条从根节点到叶节点的路径。
递归参数 current 表示当前路径已经形成的数值；当继续访问下一层节点时，将 current 更新为 current * 10 + node.val。
这样一来，沿着根到叶路径逐步向下递归的过程，就等价于按十进制方式不断拼接数字。
当到达叶节点时，current 恰好对应一条完整路径表示的数值，直接返回即可；
对于非叶节点，其贡献等于左子树返回值与右子树返回值之和。

时间复杂度：
每个节点仅访问一次，因此时间复杂度为 O(n)，递归栈空间复杂度为 O(h)。
"""
from collections import deque


class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def build_tree(level_order):
    if not level_order:
        return None

    nodes = [None if value is None else TreeNode(value) for value in level_order]
    children = deque(nodes[1:])

    for node in nodes:
        if node is not None:
            if children:
                node.left = children.popleft()
            if children:
                node.right = children.popleft()

    return nodes[0]


def sum_root_to_leaf_numbers(root):
    def dfs(node, current):
        if node is None:
            return 0

        current = current * 10 + node.val

        if node.left is None and node.right is None:
            return current

        return dfs(node.left, current) + dfs(node.right, current)

    return dfs(root, 0)


root1 = build_tree([1, 2, 3])
print(sum_root_to_leaf_numbers(root1))

root2 = build_tree([4, 9, 0, 5, 1])
print(sum_root_to_leaf_numbers(root2))
# your algorithm time complexity is: O(n)
````
